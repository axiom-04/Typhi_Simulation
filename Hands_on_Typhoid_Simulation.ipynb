**Prompt-3:** <br>
now show a graph with 5 line graphs for 5 different mean and standard deviation using normal distribution

````python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-5, 13, 1000)  # This range should cover all the curves well

# Define the means and standard deviations
means = [0, 2, 4, 6, 8]
std_devs = [1, 0.5, 0.2, 1.5, 0.3]

# Plot each normal distribution curve
for mean, std_dev in zip(means, std_devs):
    y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)
    plt.plot(x, y, label=f"μ={mean}, σ={std_dev}")

# Add legend, title, and labels
plt.legend()
plt.title("Normal Distributions with Different Means and Standard Deviations")
plt.xlabel("x")
plt.ylabel("Probability Density")

plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define n and p parameter combinations
params = [
    (20, 0.5),
    (30, 0.3),
    (40, 0.7),
    (50, 0.2),
    (60, 0.6)
]

# Set the figure size
plt.figure(figsize=(12, 8))

# Plot binomial PMF for each n and p combination
for n, p in params:
    x = np.arange(0, n+1)
    y = binom.pmf(x, n, p)
    plt.plot(x, y, '-o', label=f'n={n}, p={p}')

# Set labels, title, and legend
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF for Different n and p Values')
plt.legend()

# Display the plot
plt.show()
````

**Prompt-3:** <br>
show me 5 different line graphs for 5 lamdas in the same graph using poisson distribution

````python
import numpy as np
import matplotlib.pyplot as plt

# Setting the random seed for reproducibility
np.random.seed(0)

# Define the PMF of Poisson distribution
def poisson_pmf(k, lmbda):
    return (np.exp(-lmbda) * lmbda**k) / np.math.factorial(k)

# Lambda values
lambdas = [2, 4, 6, 8, 10]

# Generate k values (e.g., from 0 to 20 for clear visualization)
k_values = np.arange(0, 21)

# Plot the PMF for each lambda value
for lmbda in lambdas:
    probabilities = [poisson_pmf(k, lmbda) for k in k_values]
    plt.plot(k_values, probabilities, label=f'λ={lmbda}', marker='o')

plt.title('Poisson Distribution PMF for Different λ Values')
plt.xlabel('k')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
````

##Confidence Intervals

**Prompt-1:** <br>
why do we use confidence interval? explain me with an easy example

**Prompt-2:** <br>
how to calculate ci in python

````python
import numpy as np
from scipy import stats

# Sample data
sample_mean = 99
sample_std_dev = 3
n = 30
confidence_level = 0.95

# Calculate the standard error
standard_error = sample_std_dev / np.sqrt(n)

# Calculate the confidence interval
ci = stats.t.interval(
    confidence_level,        # Confidence level
    df=n-1,                  # Degrees of freedom
    loc=sample_mean,         # Sample mean
    scale=standard_error     # Standard error
)

print("95% Confidence Interval:", ci)
````

**Prompt-3:** <br>
Relative Half-Width of 95% CI

````python
import numpy as np
from scipy import stats

# Sample data
sample_mean = 99
sample_std_dev = 3
n = 30

# Calculate the standard error
standard_error = sample_std_dev / np.sqrt(n)

# Calculate the half-width of the 95% CI using the t-distribution
alpha = 0.05 # for a 95% CI
t_value = stats.t.ppf(1 - alpha/2, df=n-1)
half_width = t_value * standard_error

# Calculate the relative half-width
relative_half_width = half_width / sample_mean

print("Relative Half-Width:", relative_half_width)
````

**Prompt-4:** <br>
take no of cases 2500, 3 probabilities such as 0.1, 0.25, 0.5 and 5 iterations with no of simulations (200, 500, 1000, 2000, 5000, 10000). for each probability and iteration generate random numbers using binomial distribution individually for size =no of simulations. then calculate the confidence interval and Relative Half-Width of 95% CI for each probability and no of simulations

````python
import numpy as np
from scipy import stats

# Define the probabilities and sizes
probabilities = [0.1, 0.25, 0.5]
sizes = [200, 500, 1000, 2000, 5000, 10000]
iterations = 5

# Function to calculate the confidence interval
def calculate_CI(data, alpha=0.05):
    sample_mean = np.mean(data)
    standard_error = np.std(data, ddof=1) / np.sqrt(len(data))
    t_value = stats.t.ppf(1 - alpha/2, df=len(data)-1)
    half_width = t_value * standard_error
    return (sample_mean - half_width, sample_mean + half_width), half_width / sample_mean

for p in probabilities:
    print(f"Probability: {p}")
    for n in sizes:
        print(f"\tSize: {n}")
        for i in range(iterations):
            # Generate random numbers using binomial distribution
            data = np.random.binomial(2500, p, n)

            # Calculate the confidence interval and relative half-width
            CI, RHW = calculate_CI(data)
            print(f"\t\tIteration {i + 1}: CI = {CI}, RHW = {RHW}")
````

**Prompt-5:** <br>
now show me 3 line plots for each probability with x axis = no of simulations and y axis = relative half width 95% ci

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Define the probabilities and sizes
probabilities = [0.1, 0.25, 0.5]
sim_sizes = [200, 500, 1000, 2000, 5000, 10000]
iterations = 5
n = 2500  # No of cases

# Function to calculate the confidence interval
def calculate_CI(data, alpha=0.05):
    sample_mean = np.mean(data)
    standard_error = np.std(data, ddof=1) / np.sqrt(len(data))
    t_value = stats.t.ppf(1 - alpha/2, df=len(data)-1)
    half_width = t_value * standard_error
    return half_width / sample_mean

results = {}

for p in probabilities:
    rhw_values = []
    for size in sim_sizes:
        avg_rhw = np.mean([calculate_CI(np.random.binomial(n, p, size)) for _ in range(iterations)])
        rhw_values.append(avg_rhw)
    results[p] = rhw_values

# Plotting the results
plt.figure(figsize=(10,6))

for p, rhw_values in results.items():
    plt.plot(sim_sizes, rhw_values, label=f'Probability {p}', marker='o')

plt.xlabel('Number of Simulations')
plt.ylabel('Relative Half-Width of 95% CI')
plt.title('Relative Half-Width of 95% CI vs. Number of Simulations')
plt.legend()
plt.grid(True)
plt.show()
````

## Simulation: Bangladesh inpatient (DSH) for typhi

The Supplementary link is
[here](https://www.thelancet.com/cms/10.1016/S2214-109X(22)00119-X/attachment/b0f16a56-80cf-4127-b0e2-1f471ee1ec69/mmc1.pdf)

For Bangladesh inpatient (DSH) the table is given below:

| Age Group | A1 | A2 | A3 | Cases| Population
|---------|---------|---------|---------|---------|---------|
| < 2 years| 0·59| 0.989|0.613|35|77,958
| 2 - 4 years| 0·59| 0.988|0.613|80|1,73,878
| 5 - 15 years| 0·59| 0.991|0.613|85|5,88,070


````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import matplotlib.pyplot as plt
#Initialization of fixed parameters
numbers = (35, 80, 85)
a1 = (0.59, 0.59, 0.59)
a2 = (0.989, 0.988, 0.991) #DSH
a3 = (0.613, 0.613, 0.613)
pop_changing = (77958, 173878, 588070)
age_group = ('<2','2-4','5-15')
n_iter = 100000

incidence = []
lower_ci = []
upper_ci = []


#Define alpha and beta values
def getAlphaBeta(mu, sigma):
    alpha=(((1-mu)/sigma**2)-(1/mu))*mu**2
    beta=((1/mu)-1)*alpha

    return alpha, beta
#Calculate adjusted incidence for each year

for x in range(len(age_group)):
  # Define the distribution of cases
  true_incidence = numbers[x]/pop_changing[x]
  cases_dist = np.random.binomial(n=pop_changing[x], p=true_incidence, size=n_iter)

  # Define the distributions of the adjustment factors
  (alpha1,beta1) = getAlphaBeta(a1[x],0.05)
  (alpha2,beta2) = getAlphaBeta(a2[x],0.05)
  (alpha3,beta3) = getAlphaBeta(a3[x],0.05)
  a1_dist = np.random.beta(alpha1, beta1, size=n_iter)
  a2_dist = np.random.beta(alpha2, beta2, size=n_iter)
  a3_dist = np.random.beta(alpha3, beta3, size=n_iter)

  # Calculate the crude incidence rate
  crude_rate_dist = cases_dist / pop_changing[x]

  # Calculate the adjusted incidence rate
  adjusted_rate_dist = crude_rate_dist / (a1_dist * a2_dist * a3_dist)
  incidence.append(np.mean(adjusted_rate_dist*100000))

  # Calculate the uncertainty intervals (e.g. 95% confidence intervals) for the adjusted incidence rate
  lower_ci.append(np.percentile(adjusted_rate_dist*100000,2.5))
  upper_ci.append(np.percentile(adjusted_rate_dist*100000,97.5))

incidence
#Plotting the values

plt.plot(age_group, incidence, '#8F262A', linewidth=2, markersize=12)
#plt.plot(year, lower_ci,'#B298CB', linewidth=2, markersize=12)
#plt.plot(year, upper_ci,'#98CB98', linewidth=2, markersize=12)
plt.fill_between(age_group, lower_ci, upper_ci, color='#8F262A', alpha=.15)
plt.legend(['Adjusted incidence'])
plt.title("simulation of typhi incidence")
````
